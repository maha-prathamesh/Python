# String Function

## Manipulation

```python
#capitalize
#Upper case the first letter in this sentence
str_ = 'arise, awake, and stop not until the goal is achieved.' 
x = str_.capitalize()
print ('capitalize op ===> ', x)


#casefold
#Converts string into lower case
str_ = 'Arise, Awake, And sTop not uNtil the goal is achieved.' 
x = str_.casefold()
print ('casefold op ===> ', x)


#center
#Returns a centered string
#Converts string into lower case
x = str_.center(60, 'P')
print('center op ===> ', x)


#join 
x = 'arise, awake, and stop not until the goal is achieved.'.join('TTT')
print('Join ===> ',x)


#lower
x = str_.lower()
print('Lower ===>', x)


#lstrip
str_ = '   arise, awake, and stop not until the goal is achieved.    '
x = str_.lstrip()
y = str_.lstrip(' arise')
print('lstrip ===>', x)
print('lstrip ===>', y)


#rstrip
str_ = '   arise, awake, and stop not until the goal is achieved.    '
x = str_.rstrip()
y = str_.rstrip('achieved.  ')
print('rstrip ===>', x)
print('rstrip ===>', y)


#maketrans
str_ = 'arise, awake, and stop not until the goal is achieved.'
x = str_.maketrans("t", "o")
print('make trans ===>', str_.translate(x))


#awake
x = str_.partition("awake")
print('Partition ===>', x)


#replace
x = str_.replace("awake", "awwwwwwww")
print('Replace ===> ', x)
```

```python
#encode
str_ = 'Ari@@@se, Awak???e, And sToåp not åuNtil the goal is achieved.' 
x = str_.encode()
print(str_.encode(encoding="ascii",errors="backslashreplace"))
print(str_.encode(encoding="ascii",errors="ignore"))
print(str_.encode(encoding="ascii",errors="namereplace"))
print(str_.encode(encoding="ascii",errors="replace"))
print(str_.encode(encoding="ascii",errors="xmlcharrefreplace"))

```

```python
#format
str_ = "Arise, {a}, And sTop not uNtil the goal is achieved.".format(a = "Awake")
print(str_)

str_ = "Arise, {0}, And sTop not uNtil the goal is achieved.".format("Awake")
print(str_)

str_ = "Arise, {}, And sTop not uNtil the goal is achieved.".format("Awake")
print(str_)

```

## Retrieve

```python
str_ = 'Arise, Awake, And sTop not uNtil the goal is achieved.' 

#count
#Returns the number of times a specified value occurs in a string
x = str_.count("Awake")
print('Awake Count in str_ ===>', x)

#endswith
x = str_.endswith(".")
print("endswith   '.'  ",x)

#find
x = str_.find("not", 0 , 30)
print("Find index of 'not' ", x)

#index
x = str_.index("not", 0, 30)
print('Index of ', x)



```

## boolean functions

```python
str_ = 'Arise'
#isalpha
print(str_.isalpha())

#isalnum
print('a1'.isalnum())

#isdecimal
print('1'.isdecimal())

#isdigit
print('1'.isdigit())

#isidentifier
print('aa'.isidentifier())

#islower
print('aa'.islower())

#isupper
print('AA'.isupper())

#isnumeric
print('1293'.isnumeric())

#isnumeric
print('aa'.isprintable())
print('Ak'.istitle())


```
