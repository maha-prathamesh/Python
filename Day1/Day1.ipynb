# Control Flow

## If

```python
x = int(input("Please enter an integer: "))

if x < 10 :
    print('Number is less than 10')
```

## If - Else

```python
x = int(input("Please enter an integer: "))

if x < 10 :
    print('Number is less than 10')
else:
    print('Number is greater than 10')
```

## If - Elif - Else 

```python
x = int(input("Please enter an integer: "))

if x < 2 :
    print('Number is less than 2')
elif x < 10:
    print('Number is less than 10')
else:
    print('Number is greater than 10')
```

## break and continue Statements

```python
### Break
for n in range(2, 10):
    for x in range(2, n):
        if n % x == 0:
             print(n, 'equals', x, '*', n//x)
             break
        else:
         # loop fell through without finding a factor
         print(n, 'is a prime number')
```

```python
### Continue
for num in range(2, 10):
    if num % 2 == 0:
        print("Found an even number", num)
        continue
        print("Found an odd number", num)
```

# Type Casting

```python
a = 10.0
print(int(a))

b = 10
print(float(b))

s = '10'
print(int(s))

s1 = '10.0'
print(float(s1))


print(str(float(s1)))
print(type(str(float(s1))))

s3 = "this is type casting"
print(int(s3))
```
